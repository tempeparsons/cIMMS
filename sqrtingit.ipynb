```python
import numpy as np
import matplotlib.pyplot as plt

filename_list = ['2021-03-03_IAPP_2n5_sliceA_0V_dt_extracted.txt', '2021-03-03_IAPP_2n5_sliceA_10V_dt_extracted.txt', 
                 '2021-03-03_IAPP_2n5_sliceA_20V_dt_extracted.txt', '2021-03-03_IAPP_2n5_sliceA_30V_dt_extracted.txt', 
                 '2021-03-03_IAPP_2n5_sliceA_40V_dt_extracted.txt', '2021-03-03_IAPP_2n5_sliceA_50V_dt_extracted.txt', 
                 '2021-03-03_IAPP_2n5_sliceA_60V_dt_extracted.txt', '2021-03-03_IAPP_2n5_sliceA_70V_dt_extracted.txt']

XData = []
YData = []

for filename in filename_list:
    filenameparts = filename.split('_')
    voltage = filenameparts[4].replace('V', '')
    voltage = int(voltage)
    arrtime = []
    intensity = []
    xdata = []
    ydata = []
    with open (filename, 'r') as f:
        for dataline in f:
            dataline = dataline.replace('\n', '')
            arrtime_intensity = dataline.split('\t')
            arrtime.append(float(arrtime_intensity[0]))
            intensity.append(int(arrtime_intensity[1]))
        #print(len(intensity))
        #print(sum(i > 0 for i in intensity))
    for index, intensity_value in enumerate(intensity):
        if intensity_value > 0.0:
            n = int(np.log2(intensity_value)*10)
            #n = int(intensity_value/100.)
            xdata += [arrtime[index]]*n
    
    ydata = [voltage]*len(xdata)
   # print(len(xdata))
   # print(len(ydata))
    XData.extend(xdata)
    YData.extend(ydata)
    arrtime = []
    intensity = []
    xdata = []
    ydata = []
    
#print(len(XData))
#print(len(YData))

import numpy as np
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture

XArr = np.array(XData)
YArr = np.array(YData)
XYArr = np.column_stack((XArr,YArr))
```

```python
fig, ax = plt.subplots(1,4)
fig.set_size_inches(18,2)

plt.subplot(1,4,1)
gm3 = GaussianMixture(n_components=3, covariance_type='full', max_iter=50, random_state=1).fit(XYArr)
labels = gm3.predict(XYArr)
plt.scatter(XArr, YArr, c=labels, cmap = 'viridis', s=24, alpha=0.02)

plt.subplot(1,4,2)
gm3 = GaussianMixture(n_components=3, covariance_type='full', max_iter=50, init_params='k-means++', random_state=1).fit(XYArr)
labels = gm3.predict(XYArr)
plt.scatter(XArr, YArr, c=labels, cmap = 'viridis', s=24, alpha=0.02)

plt.subplot(1,4,3)
gm3 = GaussianMixture(n_components=3, covariance_type='full', max_iter=50, init_params='random', random_state=1).fit(XYArr)
labels = gm3.predict(XYArr)
plt.scatter(XArr, YArr, c=labels, cmap = 'viridis', s=24, alpha=0.02)

plt.subplot(1,4,4)
gm3 = GaussianMixture(n_components=3, covariance_type='full', max_iter=50, init_params='random_from_data', random_state=1).fit(XYArr)
labels = gm3.predict(XYArr)
plt.scatter(XArr, YArr, c=labels, cmap = 'viridis', s=24, alpha=0.02)
```
